```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel('G:\my trials\Telco-Customer-Churn.xlsx')
df
```

```python
df.columns
```

# Compare churn count with respect to gender

```python
sns.countplot(x='gender',hue='Churn',data=df)
plt.title(' Churn count vs Gender participated')
plt.xlabel('Gender')
plt.ylabel('Churn') 
```

# FIND OUT HOW MANY FEMALE SENIOR CITIZENS IN THE DATASET

```python
df1=df[df['gender']=='Female']
df2=df1[df1['SeniorCitizen']==1]
df2
```

```python
df2.SeniorCitizen.count()
df2.gender.count()
```

INFERENCE:A TOTAL OF 568 SENIOR  FEMALE CITIZENS ARE PRESENT IN DATASET.

# Compare 'tenure' with 'Total Charges'

```python
plt.figure(figsize=(15,5))
df['TotalCharges']=pd.to_numeric(df.TotalCharges,errors='coerce')
plt.bar(df['tenure'],df['TotalCharges'])
plt.title(' Tenure vs Total charges')
plt.xlabel('Tenure')
plt.ylabel('TotalCharges') 
```

# Findout which contract preffered by the senior citizen.

```python
df1=df['Contract'].value_counts().to_frame().reset_index()
df1
```

```python
df2=df['SeniorCitizen'].value_counts().to_frame().reset_index()
df2
```

```python
sns.countplot(x='Contract',hue='SeniorCitizen', data=df)
plt.title('Contract preferred by Citizens')
plt.grid()
plt.show()
```

# Comment your finds on Payment Method?

```python
df1=df['PaymentMethod'].value_counts().to_frame().reset_index()
df1
```

```python
plt.figure(figsize=(12,5))
plt.bar(df1['index'],df1['PaymentMethod'])
for i,j in zip(df1['index'],round(df1['PaymentMethod'],2)):
 plt.annotate(str(j),(i,j+2))
plt.title(' Different payment methods ')
plt.xlabel('Payment Methods')
plt.ylabel('Count') 
```
